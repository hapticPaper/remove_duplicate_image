```python
import hashlib, os
file_list = []
def hashdir(directory):
    duplicates = []
    hash_keys = dict()
    for index, entry in enumerate(os.listdir(directory)):  #listdir('.') = current directory
        if os.path.isfile(entry):
            file_list
            with open(entry, 'rb') as f:
                filehash = hashlib.md5(f.read()).hexdigest()
            if filehash not in hash_keys: 
                hash_keys[filehash] = index
            else:
                duplicates.append((index,hash_keys[filehash]))
        else:
            hashdir(os.path.join(directory, entry))
```

```python
duplicates
```

```python
for file_indexes in duplicates:
    try:
    
        plt.subplot(121),plt.imshow(Image.open(file_list[file_indexes[1]]))
        plt.title(file_indexes[1]), plt.xticks([]), plt.yticks([])

        plt.subplot(122),plt.imshow(Image.open(file_list[file_indexes[0]]))
        plt.title(str(file_indexes[0]) + ' duplicate'), plt.xticks([]), plt.yticks([])
        plt.show()
    
    except OSError as e:
        continue
```

# Delete Files After Printing

```python
for index in duplicates:
    os.remove(file_list[index[0]])
```
